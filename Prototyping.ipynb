```python
import matplotlib.pyplot as plt
import numpy as np

# Function to draw a single hexagon
def draw_hexagon(ax, center, size, gap, edge_color='green', face_color='white'):
    x, y = center
    angle = np.linspace(0, 2 * np.pi, 7)  # 7 points to close the hexagon
    x_hex = x + (size - gap) * np.cos(angle)
    y_hex = y + (size - gap) * np.sin(angle)
    ax.fill(x_hex, y_hex, facecolor=face_color, edgecolor=edge_color, linewidth=1)

# Parameters
size = 1  # Radius of hexagon
gap = 0.1  # Gap between tiles
rows, cols = 5, 4  # Number of rows and columns of hexagons

# Create the plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.set_axis_off()
ax.set_facecolor('white')

# Generate hexagonal grid
for row in range(rows):
    for col in range(cols):
        x_offset = 1.5 * size * col
        y_offset = np.sqrt(3) * size * row
        if col % 2 == 1:  # Offset for alternate rows
            y_offset += np.sqrt(3) * size / 2
        draw_hexagon(ax, (x_offset, y_offset), size, gap)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to draw a single hexagon
def draw_hexagon(ax, center, size, edge_color, face_color, linewidth=1, angle=0):
    x, y = center
    angles = np.linspace(0, 2 * np.pi, 7) + np.radians(angle)  # Apply rotation
    x_hex = x + size * np.cos(angles)
    y_hex = y + size * np.sin(angles)
    ax.fill(x_hex, y_hex, facecolor=face_color, edgecolor=edge_color, linewidth=linewidth)

# Parameters
tile_size = 1       # Radius of green tiles
gap = 0.2           # Gap between green tiles
outer_size = tile_size + gap  # Radius of gray hexagons
rows, cols = 3, 3   # Number of rows and columns (for a small visible grid)

# Offset inputs for the central green tile
x_offset = 0.2     # Offset in the x-direction
y_offset = -0.1    # Offset in the y-direction
angle_offset = 15  # Rotation angle in degrees

# Create the plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.set_axis_off()
ax.set_facecolor('white')

# Generate a tiled hexagonal grid
for row in range(-1, rows + 1):  # Extend to ensure surrounding edges
    for col in range(-1, cols + 1):
        # Calculate hexagon center positions
        x_base = col * 1.5 * outer_size
        y_base = row * np.sqrt(3) * outer_size
        if col % 2 == 1:  # Offset for alternate rows
            y_base += np.sqrt(3) * outer_size / 2

        # Draw the outer (gray) hexagon in its original position
        draw_hexagon(ax, (x_base, y_base), outer_size, 'lightgray', 'none', linewidth=0.5)

        # Draw the inner (green) hexagon
        if (x_base, y_base) == (0, 0):  # Central hexagon with offsets
            draw_hexagon(
                ax,
                (x_base + x_offset, y_base + y_offset),
                tile_size,
                'green',
                'white',
                linewidth=1.5,
                angle=angle_offset
            )
        else:  # Neighboring hexagons
            draw_hexagon(ax, (x_base, y_base), tile_size, 'green', 'white', linewidth=1.5)

# Set a closer zoom for the central hexagon and immediate surroundings
zoom_extent = 1.5 * outer_size
ax.set_xlim(-zoom_extent, zoom_extent)
ax.set_ylim(-zoom_extent, zoom_extent)

# Show the plot
plt.show()
```
